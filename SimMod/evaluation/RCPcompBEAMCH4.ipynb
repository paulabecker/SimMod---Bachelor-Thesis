```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import plot functions
from utils import load_data, lineplot, lineplots_on_grid

TAG45 = "200810-193117-BEAM-rcp45"

# Read in the pd.DataFrame
data45 = load_data(TAG45)

# Import RCP data
data45_ref = pd.read_table("../results/RCP45_MIDYR_CONC.DAT",
                      skiprows=38, sep='\s+')
```

```python
data45_ref
```

```python
# choose RCP data
keep_ref = data45_ref[data45_ref['YEARS']>1999]
keep_ref = keep_ref[keep_ref['YEARS']<2100]
time_ref45 = keep_ref['YEARS']
ch4_ref45 = keep_ref['CH4']
```

```python
# choose SimMod data
keep = data45[data45['year']>1999]
time45 = keep['year']
ch445 = keep['ch4_ppb']
```

```python
# SimMod GHG concentrations
plt.plot(time45, ch445, ls = 'dashed')

# RCP GHG concentrations
plt.plot(time_ref45, ch4_ref45)

plt.title("CH4 concentration SimMod and RCP for RCP 4.5")
plt.xlabel("year")
plt.ylabel("ppb Atsmospheric CH4")
```
