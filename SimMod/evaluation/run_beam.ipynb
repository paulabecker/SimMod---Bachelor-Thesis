```python
from beam_carbon.beam import BEAMCarbon
import numpy as np
import pandas as pd
```

```python
beam = BEAMCarbon()
#beam._k_1 = 2.
print(beam.k_1) 
beam.time_step = 10.
beam.intervals = 200
beam.emissions = np.array([
    9.58, 12.25, 14.72, 16.07, 17.43, 19.16, 20.89, 23.22, 26.15, 29.09
])
beam.temperature_dependent = False
beam.linear_temperature = True
results = beam.run()

results
```

```python
beam = BEAMCarbon()
beam.time_step = 1.
beam.intervals = 3
beam.emissions = np.array([10.,10.,20.,20.
])
beam.temperature_dependent = False
beam.linear_temperature = False
results1 = beam.run()

results1
```

```python
results1_comp = results1.loc[['mass_atmosphere', 'mass_upper', 'mass_lower',
                   'temp_atmosphere', 'temp_ocean', 'phi12', 'phi22',
                   'cumulative', 'A', 'B'],[0.0,2.0,4.0]]
results1_comp
```

```python
results1_comp = results1.iloc[:,[0,2,4]]
results1_comp
```

```python
beam = BEAMCarbon()
beam.time_step = 2.
beam.intervals = 6
beam.emissions = np.array([10.,20.
])
beam.temperature_dependent = False
beam.linear_temperature = False
results2 = beam.run()

results2
```

```python
results1_comp.equals(results2)
```

```python
resultsBool = (results1_comp != results2).stack()  # Create Frame of comparison booleans
resultsdiff = pd.concat([results1_comp.stack()[resultsBool], results2.stack()[resultsBool]], axis=1)
resultsdiff.columns=["results1", "results2"]
print(resultsdiff)
```
