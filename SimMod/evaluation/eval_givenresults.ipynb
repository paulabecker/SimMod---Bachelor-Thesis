```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

FILE_NAME = "simmod_run_8p5_pulse_response"
#FILE_NAME = "simmod_run_4.5 pulse response"
```

```python
data = pd.read_csv("../given_results/" + FILE_NAME + ".csv", sep=',')
prop = 'total_forcing_rcp'

year = data['year']
y = data[prop]

plt.plot(year, y)
```

```python
year = data['year']

fig, axes = plt.subplots(10, 3, figsize=(13,40))

for i, col in enumerate(data.columns[3:]):
    y = data[col]
    
    axes[i//3, i%3].plot(year, y)
    axes[i//3, i%3].set_xlabel('year')
    axes[i//3, i%3].set_ylabel(col)
```
