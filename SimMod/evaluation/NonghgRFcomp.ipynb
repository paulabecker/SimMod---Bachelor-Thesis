```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import plot functions
from utils import load_data, lineplot, lineplots_on_grid
```

```python
# Read in the pd.DataFrame
TAG45 = "200809-183506-pulse_response-rcp45"
data45 = load_data(TAG45)

# Import RCP data
data45_ref = pd.read_table("../results/RCP45_MIDYEAR_RADFORCING.DAT",
                      skiprows=59, sep='\s+')
```

```python
# choose CMIP5 data
keep_ref = data45_ref[data45_ref['YEARS']<2100]
time_ref45 = keep_ref['YEARS']
rf_ghg_ref45 = keep_ref['TOTAL_ANTHRO_RF'] - keep_ref['CO2CH4N2O_RF']

# choose SimMod data
time45 = data45['year']
rf_ghg45 = data45['rcp_nonghg_forcing']
```

```python
plt.plot(time45, rf_ghg45, ls = 'dashed', c = 'r')
plt.plot(time_ref45, rf_ghg_ref45, c= 'r')

plt.title("SimMod: rcp nonghg forcing , CMIP5: total anthropogenic RF - GHG forcing")
plt.xlabel("year")
plt.ylabel("W/m^2")
```
