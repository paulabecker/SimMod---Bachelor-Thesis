__Plotting script visualizing the hysteresis effect for the Implied TCR.__

```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import pandas as pd
%matplotlib inline
plt.rcParams['font.size'] = 15
```

```python
dpi = 100
```

```python
simmod_path = "../data/preprocessed/SimMod_timeseries.csv"

simmod = pd.read_csv(simmod_path)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

plot26 = ax.scatter(simmod['SimMod2.6_f'], simmod['SimMod2.6_t']+0.2,
                    c=simmod['year'], s=24, cmap='turbo')
plot45 = ax.scatter(simmod['SimMod4.5_f'], simmod['SimMod4.5_t'],
                    c=simmod['year'], s=24, cmap='turbo')
plot60 = ax.scatter(simmod['SimMod6.0_f'], simmod['SimMod6.0_t']-0.2,
                    c=simmod['year'], s=24, cmap='turbo')
#ax.scatter(simmod['SimMod8.5_f'], simmod['SimMod8.5_t'], c=simmod['year'],
#            s=15, cmap='plasma', label='RCP 8.5')

ax.annotate("RCP 2.6", (-1.8,-0.45), size=13)
ax.annotate("RCP 4.5", (-1.8,-0.65), size=13)
ax.annotate("RCP 6.0", (-1.8,-0.85), size=13)

ax_inset = fig.add_axes([0.165, 0.65, 0.22, 0.2])
ax_inset.plot(simmod['year'], simmod['SimMod2.6_f'], c='k', lw=2)
ax_inset.plot(simmod['year'], simmod['SimMod4.5_f'], c='k', lw=2)
ax_inset.plot(simmod['year'], simmod['SimMod6.0_f'], c='k', lw=2)
ax_inset.set_xlim(1760, 2500)
ax_inset.tick_params(direction='in', labelleft=False)
ax_inset.set_ylabel(r"$\Delta F$")

fig.colorbar(plot60, ax=ax, aspect=30, pad=0.03)
ax.set_xlim(-2, 4.5)
ax.grid()
ax.set_axisbelow(True)
ax.set_xlabel(r"$\Delta F$ $[Wm^{-2}]$")
ax.set_ylabel(r"$\Delta T$ $[\degree C]$")
#plt.legend(scatterpoints=3)
#fig.savefig("figures/SimMod_TCR_hysteresis.pdf", dpi=dpi, bbox_inches="tight")
plt.show()
```
